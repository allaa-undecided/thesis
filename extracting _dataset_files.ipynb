```python
import os

# get all files in path

files = []

for file in os.listdir():
    print(f"Read file: {file}")
    files.append(file)

print(files)
```

```python

for file in files:
    if file.endswith(".rar"):
        get_ipython().system_raw(f"unrar x {file}")
```

```python
# change dir names

paths = []
for path in os.listdir():
    if os.path.isdir(path):
        paths.append(path)

print(paths)
```

```python
# rename folders

for path in paths:
    os.rename(path, path.replace(" ", ""))
```

```python
import shutil
```

```python
# combine them all in one folder

source_folders = [path.replace(" ", "") for path in paths]

dest_folder = "dataset"

# check if folder present

if not os.path.exists(dest_folder):
    os.mkdir(dest_folder)

i = 0
for src in source_folders:
    for root, _, files in os.walk(src):
        for file in files:
            # join
            src_file_path = os.path.join(root, file)
            dest_file_path = os.path.join(dest_folder, file)
            print(f"copying {src_file_path} to {dest_file_path}")
            i+=1
            shutil.copy2(src_file_path, dest_file_path)

print(f"i : {i}")
```
